```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame
data = {
    "Gender": ["Male", "Female", "Male", "Female", "Male", "Female", "Male", "Female"],
    "Promo Code Used": ["Yes", "No", "No", "Yes", "Yes", "No", "Yes", "Yes"]
}
df = pd.DataFrame(data)

# Step 1: Compute counts
counts = df.groupby(["Gender", "Promo Code Used"]).size().unstack(fill_value=0)

# Step 2: Calculate percentages
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

# Step 3: Create stacked bar chart
fig, ax = plt.subplots(figsize=(8, 6))
bottoms = [0] * len(percentages)  # Initialize the bottom for stacking

for promo_status in percentages.columns:
    ax.bar(percentages.index, percentages[promo_status], label=promo_status, bottom=bottoms)
    bottoms += percentages[promo_status]

# Step 4: Add percentage labels
for i, gender in enumerate(percentages.index):
    cumulative = 0
    for promo_status in percentages.columns:
        value = percentages.loc[gender, promo_status]
        if value > 0:  # Add label only if value > 0
            ax.text(i, cumulative + value / 2, f'{value:.1f}%', ha='center', va='center', fontsize=10)
        cumulative += value

# Customize the plot
ax.set_title("Promo Code Usage by Gender (as % of total)", fontsize=14)
ax.set_xlabel("Gender", fontsize=12)
ax.set_ylabel("Percentage (%)", fontsize=12)
ax.legend(title="Promo Code Used")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Example DataFrame (replace this with your actual DataFrame)
# Assuming 'df' is already loaded and contains the relevant columns
# df = pd.read_csv('your_data.csv')

# List of columns you want to plot
columns = ['Gender', 'Subscription Status', 'Payment Method',
           'Shipping Type', 'Discount Applied', 'Promo Code Used',
           'Item Purchased','Size', 'Season']

# Set up the figure with subplots (you can adjust the grid size based on the number of plots)
fig, axes = plt.subplots(3, 3, figsize=(15, 15))  # 3x3 grid, adjust size as needed

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Loop through each column and plot a count plot
for i, column in enumerate(columns):
    sns.countplot(x=column, data=df, ax=axes[i],hue = df["Category"])
    axes[i].set_title(f'Count Plot: {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Count')

# Hide the last unused subplot (if any)
if len(columns) < len(axes):
    axes[len(columns)].axis('off')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

#Each count plot gives you a visual distribution of the values in the categorical columns, helping you understand how the data is distributed across different categories.

#The tight_layout() function will ensure that each plot is clearly visible without overlap or misalignment.

#The code also takes care of the unused subplot by hiding it, ensuring that the grid of subplots looks neat even though there are fewer plots than grid cells.
